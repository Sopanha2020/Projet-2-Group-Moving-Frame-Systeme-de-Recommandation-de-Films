# TMDB Source

```python
import gzip
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df_tmdb_new = pd.read_csv('tmdb_full.csv') 
df_tmdb_new.head()
```

```python
df_tmdb = df_tmdb_new.copy()
```

```python
df_tmdb.columns
```

```python
df_tmdb = df_tmdb[["imdb_id",'title','release_date','original_language', 'genres','runtime','vote_average','overview']]
df_tmdb['year'] = pd.to_datetime(df_tmdb['release_date']).dt.year
df_tmdb['year'] = df_tmdb['year'].astype('Int64')
df_tmdb['month'] = pd.to_datetime(df_tmdb['release_date']).dt.month
df_tmdb['month'] = df_tmdb['month'].astype('Int64')
df_tmdb = df_tmdb[(df_tmdb['year'] >= 1975) & (df_tmdb['year'] < 1995)]
df_tmdb = df_tmdb[(df_tmdb['vote_average'] >= 6) & (df_tmdb['vote_average'] < 10)]
df_tmdb = df_tmdb[(df_tmdb['runtime'] > 80) & (df_tmdb['runtime'] < 240)]
df_tmdb = df_tmdb[(df_tmdb['original_language'] == 'fr') | (df_tmdb['original_language'] == 'en')]
df_tmdb = df_tmdb[(df_tmdb["genres"].str.contains("Comedy")) | (df_tmdb["genres"].str.contains("Family")) | (df_tmdb["genres"].str.contains("Drama"))]
df_tmdb.head()
```

# IMDB Source

```python
df_name_basics_new = pd.read_csv('name.basics.tsv.gz', compression='gzip', delimiter='\t') 
df_name_basics_new.head()
```

```python
df_name_basics = df_name_basics_new.copy()
```

```python
df_name_basics = df_name_basics[["nconst", "primaryName", "primaryProfession", "knownForTitles"]]
df_name_basics = df_name_basics[(df_name_basics["primaryProfession"].str.contains("actor")) | (df_name_basics["primaryProfession"].str.contains("actress"))]
df_name_basics.head()
```

```python
df_title_principals_new = pd.read_csv('title.principals.tsv.gz', compression='gzip', delimiter='\t') 
df_title_principals_new.head()
```

```python
df_title_principals = df_title_principals_new.copy()
df_title_principals = df_title_principals.drop(['category','job','characters'], axis=1)
```

```python
df_imdb_actor = pd.merge(df_title_principals, df_name_basics, how="left", left_on='nconst', right_on='nconst')
df_imdb_actor = df_imdb_actor.dropna()
df_imdb_actor.head()
```

```python
df_imdb_actor = df_imdb_actor[(df_imdb_actor["primaryProfession"].str.contains("actor")) | (df_imdb_actor["primaryProfession"].str.contains("actress"))]
df_imdb_actor = df_imdb_actor[~df_imdb_actor.isin(['\\N']).any(axis=1)]
df_imdb_actor = df_imdb_actor[(df_imdb_actor['ordering'] == 1) | (df_imdb_actor['ordering'] == 2)]
df_imdb_actor = df_imdb_actor.drop(['nconst','ordering','knownForTitles'], axis=1)
df_imdb_actor.head()
```

```python
df_title_akas_new = pd.read_csv('title.akas.tsv.gz', compression='gzip', delimiter='\t') 
df_title_akas_new.head()
```

```python
df_title_akas = df_title_akas_new.copy()
```

```python
df_title_akas = df_title_akas[["titleId", "title", "region"]]
df_title_akas = df_title_akas[(df_title_akas['region'] == 'FR') | (df_title_akas['region'] == 'US')]
df_title_akas = df_title_akas[~df_title_akas['region'].str.contains(r'\\N', na=False)]
df_title_akas = df_title_akas[df_title_akas['region'].notna()]
df_title_akas.head()
```

```python
df_imdb_actor = pd.merge(df_title_akas, df_imdb_actor, how="left", left_on='titleId', right_on='tconst')
df_imdb_actor = df_imdb_actor.dropna()
df_imdb_actor = df_imdb_actor.drop('titleId', axis=1)
df_imdb_actor.head()
```

```python
df_imdb_actor.info()
```

```python
df_tmdb = pd.merge(df_tmdb, df_imdb_actor, how="left", left_on='imdb_id', right_on='tconst')
df_tmdb = df_tmdb.drop(['tconst','year','month'], axis=1)
df_tmdb = df_tmdb.dropna()
df_tmdb.head()
```

```python
df_tmdb.info()
```

```python
df_tmdb = df_tmdb.drop(['title_y', 'original_language'], axis=1)
df_tmdb = df_tmdb.rename(columns={'title_x': 'title'})
df_tmdb = df_tmdb.drop_duplicates()
df_tmdb.head()
```

```python
df_tmdb.info()
```

```python
df_tmdb.to_csv("filtered_tmdb_data.csv", index=False) 
```
